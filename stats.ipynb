```python
import pandas as pd
```

```python
users_visits = pd.read_excel('user_visits.xlsx')
relevant_data = ["User ID", "First Name", "Last Name", "First Name","Visited On"]
users_visits = users_visits[users_visits.columns.intersection(relevant_data)]
users_visits = users_visits.reset_index()
```

```python
users = users_visits.drop_duplicates("User ID")
users_table_columns = ["User ID", "First Name", "Last Name"]
users = users[users.columns.intersection(users_table_columns)]
users = users.reset_index()
```

```python
pd.options.mode.copy_on_write = True
october_data = users_visits[(users_visits["Visited On"] >= "2025-10-01") & (users_visits["Visited On"] <= "2025-10-31")]
october_data["Visits"] = october_data.groupby("User ID")["User ID"].transform("count")
october_data = october_data.drop_duplicates("User ID")
```

```python
sorted_data = october_data.sort_values("Visits",ascending=False)
sorted_data.to_csv("Top_visitors.csv",index=False,columns=["User ID","First Name","Last Name","Visits"])
```
