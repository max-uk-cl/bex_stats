```python
import pandas as pd
pd.options.mode.copy_on_write = True
```

```python
initial_date = "2026-01-01"
final_date = "2026-01-31"
month = "january_2026"
```

```python
users_visits = pd.read_excel('user_visits.xlsx')
relevant_data = ["User ID", "Club ID", "Club Title", "First Name", "Last Name", "First Name","Visited On"]
users_visits = users_visits[users_visits.columns.intersection(relevant_data)]
users_visits = users_visits.reset_index()
```

```python
users = users_visits.drop_duplicates("User ID")
users_table_columns = ["User ID", "First Name", "Last Name"]
users = users[users.columns.intersection(users_table_columns)]
users = users.reset_index()

clubs = users_visits.drop_duplicates("Club ID")
clubs_table_columns = ["Club ID", "Club Title"]
clubs = clubs[clubs.columns.intersection(clubs_table_columns)]
clubs = clubs.reset_index()
```

```python
monthly_data = users_visits[(users_visits["Visited On"] >= initial_date) & (users_visits["Visited On"] <= final_date)]
monthly_data["Visits"] = monthly_data.groupby("User ID")["User ID"].transform("count")
monthly_data = monthly_data.drop_duplicates("User ID")
```

```python
monthly_data_clubs = users_visits[(users_visits["Visited On"] >= initial_date) & (users_visits["Visited On"] <= final_date)]
monthly_data_clubs["Visits"] = monthly_data_clubs.groupby("Club ID")["Club ID"].transform("count")
monthly_data_clubs = monthly_data_clubs.drop_duplicates("Club ID")
```

```python
sorted_data = monthly_data.sort_values("Visits",ascending=False)
sorted_data.to_csv(month+"_top_visitors.csv",index=False,columns=["User ID","First Name","Last Name","Visits"])
```

```python
sorted_data_clubs = monthly_data_clubs.sort_values("Visits",ascending=False)
sorted_data_clubs.to_csv(month+"_top_clubs.csv",index=False,columns=["Club ID","Club Title","Visits"])
```
